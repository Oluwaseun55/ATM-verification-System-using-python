# ATM PIN Verification System: Simulate a real ATMâ€™s security. Your program should store a correct PIN and then give the user only three attempts to enter the correct one. Use a loop to control the number of attempts and an if statement inside the loop to check the PIN. If the user fails all three attempts, the account should be temporarily locked.

```python
# simple_pin_atm.py
CORRECT_PIN = "1234"   # stored correct PIN (in real systems don't store plain text)

MAX_ATTEMPTS = 3
attempts = 0

while attempts < MAX_ATTEMPTS:
    pin = input("Enter your 4-digit PIN: ").strip()
    attempts += 1

    if pin == CORRECT_PIN:
        print("PIN accepted. Access granted.")
        break
    else:
        remaining = MAX_ATTEMPTS - attempts
        if remaining > 0:
            print(f"Incorrect PIN. You have {remaining} attempt(s) remaining.")
        else:
            print("Incorrect PIN. No attempts remaining. Account temporarily locked.")
```

```python
import gradio as gr

# Stored correct PIN (for demonstration)
CORRECT_PIN = "1234"
MAX_ATTEMPTS = 3
attempts = {"count": 0}  # using a dictionary so it retains state across runs


def verify_pin(entered_pin):
    if attempts["count"] >= MAX_ATTEMPTS:
        return "Account temporarily locked. Please try again later."

    attempts["count"] += 1

    if entered_pin.strip() == CORRECT_PIN:
        result = " PIN accepted. Access granted."
        attempts["count"] = 0  # reset after success
    else:
        remaining = MAX_ATTEMPTS - attempts["count"]
        if remaining > 0:
            result = f" Incorrect PIN. You have {remaining} attempt(s) remaining."
        else:
            result = " Incorrect PIN. No attempts remaining. Account temporarily locked."
    return result


# Gradio interface
with gr.Blocks(title="ATM PIN Verification System") as demo:
    gr.Markdown("##  ATM PIN Verification System\nEnter your 4-digit PIN below:")
    pin_input = gr.Textbox(label="Enter PIN", type="password", placeholder="e.g., 1234")
    output = gr.Textbox(label="System Response", interactive=False)
    submit_btn = gr.Button("Submit")

    submit_btn.click(fn=verify_pin, inputs=pin_input, outputs=output)

demo.launch()
```
